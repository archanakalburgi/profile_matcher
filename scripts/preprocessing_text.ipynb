```python
import pandas as pd
import string
data=pd.read_excel("all_scrapped_jobs.xlsx")
data.head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import wordnet
from nltk.corpus import stopwords
import re

def tokenize(doc):
    stop_words = stopwords.words('english')

    tokens=[token.strip() \
            for token in nltk.word_tokenize(doc.lower()) \
            if token.strip() not in stop_words and\
               token.strip() not in string.punctuation]

    token = []
    wordnet_lemmatizer = WordNetLemmatizer()
    for word in tokens:
        token.append(wordnet_lemmatizer.lemmatize(word, wordnet.VERB))

    return token
```

```python
result= data['description'].apply(lambda x:tokenize(x))
print(result)
```
